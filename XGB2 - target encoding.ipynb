# Feature Engineering

```python
stock_df['Price Change %'] = ((stock_df['Current Price'] - stock_df['Start Price']) / stock_df['Start Price']) * 100
stock_df['Volatility'] = ((stock_df['Year High'] - stock_df['Year Low']) / stock_df['Year Low']) * 100
stock_df['Market Cap Rank'] = stock_df.groupby('Sector')['Market Cap'].rank(method='dense', ascending=False)
stock_df['Turnover Ratio'] = stock_df['Average Volume'] / (stock_df['Market Cap'] / stock_df['Current Price'])
```

Price Change % = (Current Priceâˆ’Start Price)/Start PriceÃ—100

Volatility = (Year Highâˆ’Year Low)/Year LowÃ—100 (Measures stock price fluctuation)

Market Cap Rank = Rank companies within the sector to compare sizes. 1 being the highest

Turnover Ratio = Average Volume/(Market Cap/Current Price) (Measures how frequently shares are traded)

```python
stock_df
```

```python
stock_df['MA_50'] = stock_df['Current Price'].rolling(window=50, min_periods=1).mean()
stock_df['MA_200'] = stock_df['Current Price'].rolling(window=200, min_periods=1).mean()
```

**Moving Averages** (e.g., 50-day, 200-day):
Calculation: The average price of a stock over a specified period (e.g., 50 days).
Significance: Smooth out price data to identify trends.

The 50-day MA is often used for short to medium-term trends, while the 200-day MA is for long-term trends. Crossovers between these can be trading signals.

```python
def calculate_rsi(prices, period=14):
        delta = prices.diff()
        gain = delta.where(delta > 0, 0)
        loss = -delta.where(delta < 0, 0)
        avg_gain = gain.rolling(window=period, min_periods=1).mean()
        avg_loss = loss.rolling(window=period, min_periods=1).mean()
        rs = avg_gain / avg_loss
        rsi = 100 - (100 / (1 + rs))
        return rsi

stock_df['RSI'] = calculate_rsi(stock_df['Current Price'])
```

**Relative Strength** Index (RSI):
Calculation: A momentum oscillator that measures the speed and change of price movements. Typically uses a 14-day period.

Significance: Identifies overbought (RSI > 70) or oversold (RSI < 30) conditions, which can signal potential reversals.

```python
from sklearn.preprocessing import OneHotEncoder
from category_encoders import TargetEncoder

# Define columns for different encodings
onehot_cols = ['Sector', 'Index']  # One-Hot Encode (low-cardinality)
target_cols = ['Company Name', 'Industry']  # Target Encode (high-cardinality)

# ---- Target Encoding ----
target_encoder = TargetEncoder()

# Fit on training data using the target variable (Current Price)
train_inputs[target_cols] = target_encoder.fit_transform(train_inputs[target_cols], train_targets)

# Apply the same transformation to test data
test_inputs[target_cols] = target_encoder.transform(test_inputs[target_cols])

# ---- One-Hot Encoding ----
onehot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Fit and transform the training data
onehot_encoder.fit(train_inputs[onehot_cols])
encoded_cols = list(onehot_encoder.get_feature_names_out(onehot_cols))

# Add encoded columns to training & test datasets
train_inputs[encoded_cols] = onehot_encoder.transform(train_inputs[onehot_cols])
test_inputs[encoded_cols] = onehot_encoder.transform(test_inputs[onehot_cols])

```

```python
# ---- Final Features Selection ----
X_train = train_inputs[numeric_cols + encoded_cols + target_cols]
X_test = test_inputs[numeric_cols + encoded_cols + target_cols]
```

## Quick & Easy Baseline Models to Benchmark Future Models

```python
def return_mean(inputs):
    return np.full(len(inputs), stock_df['Current Price'].mean())
```

```python
train_preds = return_mean(X_train)
```

```python
train_preds
```

Let's evaluate this to using the RMSE score

```python
train_targets
```

```python
def rmse(targets, preds):
    return np.sqrt(np.mean((targets - preds) ** 2))
```

```python
rmse(train_targets, train_preds)
```

```python
rmse(return_mean(X_test), test_targets)
```

The model is off by $464.52 on average

Let's try another model, which makes a random guess between the lowest and the highest Current Price

```python
def guess_random(inputs):
    lo, hi = stock_df['Current Price'].min(), stock_df['Current Price'].min()
    return np.random.random(len(inputs)) * (hi - lo) + lo
```

```python
train_preds = guess_random(X_train)
train_preds
```

```python
rmse(train_targets, train_preds)
```

```python
rmse(guess_random(X_test), test_targets)
```

This model is much worse
